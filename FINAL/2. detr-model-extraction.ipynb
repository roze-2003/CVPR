```python
import torch
import os

print("--- Extracting and Saving Model as .pth from Kaggle Output (Corrected) ---")

FINAL_EPOCH = 9 

# IMPORTANT: This path should be correct from your last run.
input_notebook_output_path = '/kaggle/input/detr-output/' 

# Path to the checkpoint file within the mounted output directory
checkpoint_path = os.path.join(input_notebook_output_path, f'detr/outputs/ciou_experiment/checkpoint_epoch{FINAL_EPOCH}.pth')

# Desired save path in the current notebook's working directory
final_model_save_path = '/kaggle/working/detr_ciou_voc_final.pth'

# --- Extraction Logic ---
if os.path.exists(checkpoint_path):
    print(f"Loading checkpoint from: {checkpoint_path}")
    
    # THE FIX IS HERE: Add weights_only=False to tell PyTorch you trust this file.
    checkpoint = torch.load(checkpoint_path, map_location='cpu', weights_only=False)
    
    model_weights = checkpoint['model']
    
    torch.save(model_weights, final_model_save_path)
    
    print(f"\nSUCCESS: Final model weights extracted and saved to: {final_model_save_path}")
    print("You can now find this file in the 'Output' section of this notebook's Data panel.")
else:
    print(f"ERROR: Checkpoint file not found at '{checkpoint_path}'")
    print("Please double-check the 'input_notebook_output_path' and ensure it's correct.")
```
